```python
import pandas as pd
import numpy as np
```

```python
file_path = './data/'

filenames = [
    'ALL_fun_f.csv',     # 基本面数据 (Fundamental)
    'ALL_mar_f.csv',     # 市场数据 (Market)
    'ALL_sen_core_f.csv',# 情绪-核心数据 (Sentiment Core)
    'ALL_sen_sat_f.csv'  # 情绪-卫星数据 (Sentiment Satellite)
]
df_fun = pd.read_csv(file_path + filenames[0])
df_mar = pd.read_csv(file_path + filenames[1])
df_sc = pd.read_csv(file_path + filenames[2])
df_ss = pd.read_csv(file_path + filenames[3])
```

```python
# --- Standardise ---
try:
    dfs_all = [df_mar, df_fun, df_sc, df_ss]
    
    for df in dfs_all:
        df['symbol'] = df['symbol'].astype(str).str.strip()
        df['date_ny'] = pd.to_datetime(df['date_ny']).dt.strftime('%Y-%m-%d')

        if 'datetime' in df.columns:
            df['datetime'] = pd.to_datetime(df['datetime']).dt.floor('min')

except Exception as e:
    print(f"*** ERROR: {e} ***")
```

```python
keys_minute = ['datetime', 'symbol', 'date_ny']
keys_day = ['date_ny', 'symbol']

# 2. "剥离" df_sc，只保留 "新" 特征
#    获取 df_mar 中已有的所有列
cols_in_mar = set(df_mar.columns)
# 找出 df_sc 中 "新" 的列
cols_new_in_sc = set(df_sc.columns) - cols_in_mar
# 我们要保留的 = 键 + 新列
cols_to_keep_sc = keys_minute + list(cols_new_in_sc)
df_sc_clean = df_sc[cols_to_keep_sc]

print(f"从 df_sc 中剥离了 {len(set(df_sc.columns)) - len(cols_to_keep_sc)} 个重复列。")
print(f"保留了 {len(cols_new_in_sc)} 个新特征。")


# 3. "剥离" df_ss，只保留 "新" 特征
#    获取 df_mar 和 df_sc 中已有的所有列
cols_in_mar_and_sc = set(df_mar.columns).union(set(df_sc.columns))
# 找出 df_ss 中 "新" 的列
cols_new_in_ss = set(df_ss.columns) - cols_in_mar_and_sc
# 我们要保留的 = 键 + 新列
cols_to_keep_ss = keys_minute + list(cols_new_in_ss)
df_ss_clean = df_ss[cols_to_keep_ss]

print(f"从 df_ss 中剥离了 {len(set(df_ss.columns)) - len(cols_to_keep_ss)} 个重复列。")
print(f"保留了 {len(cols_new_in_ss)} 个新特征。")


# 4. "剥离" df_fun (同理，以防万一)
cols_in_merged_minute = set(df_mar.columns).union(cols_new_in_sc).union(cols_new_in_ss)
cols_new_in_fun = set(df_fun.columns) - cols_in_merged_minute
cols_to_keep_fun = keys_day + list(cols_new_in_fun)
df_fun_clean = df_fun[cols_to_keep_fun]

print(f"从 df_fun 中剥离了 {len(set(df_fun.columns)) - len(cols_to_keep_fun)} 个重复列。")
print(f"保留了 {len(cols_new_in_fun)} 个新特征。")


# --- 执行最终的干净合并 ---
print("\n--- 正在执行最终的干净合并... ---")
# 1. mar + sc_clean
final_df = pd.merge(df_mar, df_sc_clean, on=keys_minute, how='left')
# 2. (mar + sc_clean) + ss_clean
final_df = pd.merge(final_df, df_ss_clean, on=keys_minute, how='left')
# 3. (mar + sc_clean + ss_clean) + fun_clean
final_df = pd.merge(final_df, df_fun_clean, on=keys_day, how='left')

print("--- 合并完成！ ---")
```

```python
# 1. 计算每列的 null 数量
null_counts = final_df.isnull().sum()

# 2. 筛选出那些 *确实* 包含 null 值的列
columns_with_nulls = null_counts[null_counts > 0]

print("--- Null 值检查结果 (mar + sc + ss) ---")

if columns_with_nulls.empty:
    print("\n🎉 恭喜！没有任何一列有 Null 值。")
else:
    print("\n[以下列存在 Null 值 (按数量排序)]:")
    # 打印排序后的结果
    print(columns_with_nulls.sort_values(ascending=False))
```
